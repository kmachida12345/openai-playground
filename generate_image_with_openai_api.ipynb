```python
import os
os.environ['OPENAI_API_KEY'] = ''
```

```python
import openai

response = openai.Image.create(
  prompt="猫たちが祝福を贈る楽しい画像を生成してください。画像には、パーティーハットをかぶった猫たちが微笑んでいる様子や、おめでとうと書かれたバナーを持っている猫、楽器を演奏している猫などが描かれていることを確認してください。また、背景にはカラフルな風船や紙吹雪が舞っている様子を表現して、祝福ムードを盛り上げてください。この画像が、愛らしさと祝福の気持ちが伝わるものであることを忘れずに。",
  n=1,
  size="256x256"
)

print(response)
image_url = response['data'][0]['url']
```
